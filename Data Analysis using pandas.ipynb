```python
x=data['TV'].values
y=data['sales'].values
```

```python
x_mean=x.mean()
```

```python
y_mean=y.mean()
```

```python
n=0
d=0
for i in range(len(x)):
    n=n+(x[i]-x_mean)*(y[i]-y_mean)
    d=d+(x[i]-x_mean)**2
print(n)
print(d)
```

```python
m=n/d
print(m)
c=y_mean-m*x_mean
print(c)
```

```python
max_x=np.max(x)+50
min_x=np.min(x)-50
X=np.linspace(min_x,max_x)
Y=m*X+c
```

```python
plt.scatter(x,y)
plt.plot(X,Y,color='#00ff11')
plt.show()
```
